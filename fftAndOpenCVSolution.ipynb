```python
#низкие частоты с fft + opencv
o = cv2.imread (f'/content/drive/MyDrive/imh/maxresdefault.jpg', 0) 
dft=cv2.dft(np.float32(o),flags=cv2.DFT_COMPLEX_OUTPUT)
dshift = np.fft.fftshift (dft) 

rows, cols = o.shape 
crow, ccol = int (rows / 2), int (cols / 2)
mask = np.zeros ((rows, cols, 2), np.int8) 
mask1 = np.zeros ((rows, cols, 2), np.int8) 
mask [crow-30: crow + 30, ccol-30: ccol + 30] = 1 
mask1 [crow-10: crow + 10, ccol-10: ccol + 10] = 1 

md=dshift*mask
ishift=np.fft.ifftshift(md)
io=cv2.idft(ishift)
io = cv2.magnitude (io [:,:, 0], io [:,:, 1]) 

md1=dshift*mask1
ishift1=np.fft.ifftshift(md1)
io1=cv2.idft(ishift1)
io1 = cv2.magnitude(io1 [:,:, 0], io1 [:,:, 1]) 
plt.figure(figsize=(30, 30), constrained_layout=False)

plt.subplot(131)
plt.imshow(o,cmap='gray')
plt.axis('off')
plt.subplot(132)
plt.imshow(io,cmap='gray')
plt.axis('off')
plt.subplot(133)
plt.imshow(io1,cmap='gray')
plt.axis('off')
plt.show()

```

```python
#в. частоты
img = cv2.imread(f'/content/drive/MyDrive/imh/maxresdefault.jpg')
imgx = cv2.Sobel(img,cv2.CV_16S,1,0,ksize=3)
imgy = cv2.Sobel(img,cv2.CV_16S,0,1,ksize=3)
imgx_uint8 = cv2.convertScaleAbs(imgx)
imgy_uint8 = cv2.convertScaleAbs(imgy)
img = cv2.addWeighted(imgx_uint8,0.5,imgy_uint8,0.5,0)
 
```
